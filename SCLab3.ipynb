# Linear Interpolasi
nilai yang diperkirakan diantara titik titik yang sudah diketahui. Misal ada data :
5 tahun -> 110 cm
10 tahun -> 140 cm
15 tahun -> 160 cm

kita mau cari tahu tinggi anak angka 7 tahun. Kita pakai interpolasi untuk menebak tinggi umur 7 tahun berdasarkan perubahan pola pertumbuhan. Karena 7 tahun itu ada di antara 5-10 tahun, jadi batas nya pakai 5 dan 10

```python
x = [0, 1, 2]
y = [1, 3, 2]
plt.scatter(x, y, color='red', label ="Data Asli")
plt.legend()
plt.show()
```

kita coba buat scatter plot data aslinya dulu biar tau data ini halus atau tajam

```python
f = interp1d(x,y) #dia nerima input x sama y
y_ask = f(1.5) #ini buat variable yang mau di cari
print(y_ask)
```

```python
plt.figure(figsize = (10,8)) #untuk ukuran gambarnya
plt.plot(x, y, '-ob'); #untuk plotnya itu akan ada  x dan y, '-' untuk membuat garis 'o' untuk membuat titik bulat sempurna 'b' itu warna biru
plt.plot(1.5, y_ask, 'ro') #'r itu wanra merah
plt.title ('Linear Interpolasi')
plt.xlabel('x') #untuk kasih nama sumbu horizontal dengan x
plt.ylabel('y') #untuk kasih nama sumbu horizontal dengan y
plt.show()
```

```python
x = np.array([0, 1, 2, 3, 4, 5]) #input
y = np.array([1, 3, 3, 5, 2, 2]) #output
f = interp1d(x,y) #menggunakan interpolasi linear

#membuat rentang nilai x baru
x_lin = np.linspace(0,5,100) # membuat 100 titik diantara 0 sampai 5
y_lin = f(x_lin) #menghitung nilai interpolasi di titik titik x_lin

y_ask = f(3.1) #meprediksi nilai y untuk titik x = 3.1
print(y_ask) #output perkiraan nilai y pada x = 3.1

plt.scatter(x, y, color='red', label="Data Asli")
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10,8))
plt.plot(x, y, 'ob')
plt.plot(x_lin, y_lin, 'b')
plt.plot(3.1, y_ask, 'ro')
plt.title ('Linear Interpolasi at x = 3.1')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

NEWTOWN INTERPOLATION

```python
%matplotlib inline
def divided_diff(x,y):

  '''
  function to calculate the devided
  differencetable
  '''

  n = len(y)
  coef = np.zeros([n,n])
  coef[:,0] = y

  for j in range(1,n):
    for i in range(n-j):
      coef[i][j] =\
       (coef[i+1][j-1] - coef[i][j-1])
  return coef

def newton_poly(coef, x_data, x): # Move newton_poly outside of divided_diff
  n = len(x_data) - 1
  p = coef[n]
  for k in range(1, n+1):
    p = coef[n-k] + (x - x_data[n-k])*p
  return p
  '''
  evaluate the newton polynomial
  at x
  '''
```

```python
x = np.array ([-5, -1, 0, 2])
y = np.array ([-2, 6, 1, 3])

a_s = divided_diff(x,y)[0,:]

x_new = np.arange (-5, 2.1, .1)
y_new = newton_poly (a_s, x, x_new)

plt.figure (figsize = (12, 8))
plt.plot (x, y, 'bo')
plt.plot (x_new, x_new, 'go')
plt.plot (x_new, y_new)
```

**Linear Interpolation :** Ibarat kamu gambar garis lurus antara dua titik, dan kamu mau cari tahu nilai di antara dua titik itu, tinggal lihat aja posisi dari dua garis lurus itu tadi. Intinya pergerakan dia itu konstam

**Newton Interpolation:** Lebih flexible, karena curvanya akan melengkung dan dapat mengenai banyak titik. Misal kita berbicara tentang korelasi anatara umur dan tinggi. Hal kayak gini kan flutuatif, jadi kurfa yang dihasilkan akan lebih akurat

```python
#data umur dan tinggi
umur = np.array([5,10,15,20])
tinggi = np.array([110,140,160,158])
```

```python
#Membuat fungsi interpolasi linear
linear_interp = interp1d(umur, tinggi)
umur_target = 7

x_lin = np.linspace(5, 10, 100)
y_lin = linear_interp(x_lin)

y_prediction = linear_interp(umur_target)
print(y_prediction)
```

```python
plt.figure(figsize = (10,8))
plt.plot(umur, tinggi, '-ob')
plt.plot(x_lin, y_lin, 'b')
plt.plot(7, y_prediction, 'ro')
plt.title ('Linear Interpolasi at x = 7')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Modul SC SESI 3

```python
jarak = np.array ([0, 5, 10])
kecepatan = np.array ([10, 15, 20])
```

```python
linear_interp = interp1d(jarak, kecepatan)
jarak_target = 7.2

x_lin = np.linspace(min(jarak), max(jarak), 100)
y_lin = linear_interp(x_lin)

y_prediction = linear_interp(jarak_target)
print(y_prediction)
```

```python
plt.figure(figsize = (10,8))
plt.plot(jarak, kecepatan, '-ob')
plt.plot(x_lin, y_lin, 'b')
plt.plot(7.2, y_prediction, 'ro')
plt.title ('Linear Interpolasi at x = 7.2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
